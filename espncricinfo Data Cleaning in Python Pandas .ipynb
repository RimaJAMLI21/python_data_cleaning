```python
def convert_balls_faced(balls_faced):
    if isinstance(balls_faced, str):
        return balls_faced.strip('+')
    else:
        return balls_faced
    
matches["balls_faced"] = matches["balls_faced"].apply(convert_balls_faced)
```
